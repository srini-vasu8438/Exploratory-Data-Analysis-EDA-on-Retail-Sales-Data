```python
#Dropping or filling the missing values
df1=df1.dropna()
df2=df2.dropna()
```

```python
#Descriptive Statistics
#Descriptive Statistics for df1
print("\nDescriptive Statistics for Retail Sales Dataset (df1):")
print(df1.describe())
```

```python
#Calculate mean, median, mode, and standard deviation for key columns
#Dataset 1
mean_total_amount = df1['Total Amount'].mean()
median_total_amount = df1['Total Amount'].median()
mode_total_amount = df1['Total Amount'].mode()[0]
std_total_amount = df1['Total Amount'].std()

print(f"\nMean Total Amount:{mean_total_amount}")
print(f"Median Total Amount:{median_total_amount}")
print(f"Mode Total Amount:{mode_total_amount}")
print(f"Standard Deviation of Total Amount:{std_total_amount}")
```

```python
#Dataset 2
#Descriptive statistics for df2
print("\nDescriptive Statistics for Menu Dataset (df2):")
print(df2.describe())
```

```python
# Calculate mean, median, mode, and standard deviation for key columns
#Dataset 2
mean_calories = df2['Calories'].mean()
median_calories = df2['Calories'].median()
mode_calories = df2['Calories'].mode()[0]
std_calories = df2['Calories'].std()

print(f"\nMean Calories:{mean_calories}")
print(f"Median Calories:{median_calories}")
print(f"mode_calories:{mode_calories}")
print(f"Standard Deviation of Calories: {std_calories}")
```

```python
#Time Series Analysis
# Convert the 'Date' column to datetime format
df1['Date'] = pd.to_datetime(df1['Date'])
```

```python
# Set the 'Date' column as the index
df1.set_index('Date', inplace=True)
```

```python
# Calculate the total sales for each day
df1['Total Sales'] = df1['Quantity'] * df1['Price per Unit']
daily_sales = df1.resample('D').sum()['Total Sales']
```

```python
# Plot the daily sales over time
plt.figure(figsize=(12, 6))
plt.plot(daily_sales.index, daily_sales.values, marker='o')
plt.title('Daily Sales Over Time')
plt.xlabel('Date')
plt.ylabel('Total Sales')
plt.grid(True)
plt.show()
```

```python
#Customer and Product Analysis
# Customer demographics
customer_demographics = df1.groupby('Gender').agg({'Age': ['mean', 'median', 'count']})
print("\nCustomer Demographics:")
print(customer_demographics)
```

```python
# Purchasing behavior by product category
purchase_behavior = df1.groupby('Product Category').agg({'Quantity': 'sum', 'Total Amount': 'sum'})
print("\nPurchasing Behavior by Product Category:")
print(purchase_behavior)
```

```python
# Plot purchasing behavior
purchase_behavior.plot(kind='bar', figsize=(10, 5), stacked=True)
plt.title('Purchasing Behavior by Product Category')
plt.xlabel('Product Category')
plt.ylabel('Total Quantity and Amount')
plt.show()
```

```python
#Visualization
# Bar chart for top-selling products
#Dataset 1
top_selling_products = df1.groupby('Product Category')['Quantity'].sum().sort_values(ascending=False)
plt.figure(figsize=(10, 5))
sns.barplot(x=top_selling_products.index, y=top_selling_products.values)
plt.title('Top Selling Products')
plt.xlabel('Product Category')
plt.ylabel('Total Quantity Sold')
plt.xticks(rotation=0)
plt.show()
```

```python
# Select numerical columns for correlation calculation
numerical_columns = ['Age', 'Quantity', 'Price per Unit', 'Total Amount']
```

```python
# Calculate the correlation matrix
correlation_matrix = df1[numerical_columns].corr()
```

```python
# Create a heatmap for the correlation matrix
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".4f", linewidths=.8)
plt.title('Correlation Matrix Heatmap')
plt.show()
```

```python
# For demonstration, let's create a sample DataFrame
data = {
    'Category': ['Breakfast', 'Breakfast', 'Breakfast', 'Breakfast'],
    'Item': ['Egg McMuffin', 'Egg White Delight', 'Sausage McMuffin', 'Sausage McMuffin with Egg'],
    'Calories': [300, 250, 370, 450],
    'Serving Size': [4.8, 4.8, 3.9, 5.7]
}

df2 = pd.DataFrame(data)
```

```python
#FacetGrid
sns.set(style="whitegrid")
g = sns.FacetGrid(df2, col="Category", height=4, aspect=1.5)
g.map(sns.scatterplot, "Serving Size", "Calories")
g.set_titles(col_template="{col_name}")
g.fig.suptitle("Calories vs Serving Size by Category", fontsize=16)
g.fig.subplots_adjust(top=0.9)
plt.show()
```
